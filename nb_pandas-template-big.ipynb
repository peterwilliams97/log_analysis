Let's Plot Some Graphs
-----------
Let's look at the [frequency](http://pandas.pydata.org/pandas-docs/dev/timeseries.html) of log messages over the duration of all the logs.

This should give a rough indication of PaperCut usage.

```python
def plot_levels(df, description, levels, subsample=True):
    for level in levels:
        if subsample:
            bars = df[df.level==level].file.resample('1S', how='count', convention='center')
            bars = pd.rolling_sum(bars, 60, center=True)
        else:
            bars = df[df.level==level].file.resample('1min', how='count', convention='center')
        bars.plot(label='%s : %d' % (level, len(df[df.level==level])))
    title('Log counts for 1 minute intervals from %s to %s: %s' % (df.index.min(), df.index.max(), description))
    ylabel('Number of log entries/minutes')
    legend()    
```

```python
plot_levels(df, 'All logs', df.level.unique())
```

```python
def plot_period(df, start, end, frequency_minutes):
    df_period = df.between_time(start, end)
    
    for i,level in enumerate(levels):
        bars = df_period[df_period.level==level].file.resample('%dmin' % frequency_minutes, how='count')
        bars.plot(kind='bar', color=colors[i], label='%s : %d' % (level, len(df_period[df_period.level==level])))
    
    minutes = (df_period.index.max() - df_period.index.min()).total_seconds()/60
    
    title(r'''Log counts for %d minute intervals from %s to %s
    $Average=\frac{%d}{%.1f}=%d$
    ''' % (frequency_minutes, df_period.index.min(), df_period.index.max(), 
           len(df_period), minutes, int(len(df_period)/minutes)), size=15)
    ylabel('Number of log entries/minute') 
    legend()  
```

### Number of file:line entries to analyze

```python
NUM_FL = 100
```

```python
fl_rolling_sums = []
for fl,ln in fl_sorted.index.values[:NUM_FL]:
    entries = df[(df.file==fl) & (df.line==ln)]
    rolling_sum = pd.rolling_sum(entries.resample('1min', how='count', convention='center'), 60, center=True)
    rolling_sum = rolling_sum/rolling_sum.mean()
    fl_rolling_sums.append(((fl,ln), rolling_sum))    
```

### The uncorrelated log messages

```python
CORR_THRESH = 0.9
from collections import OrderedDict
uncorrelated = OrderedDict()
(fl,ln),rs = fl_rolling_sums[0]
uncorrelated[(fl,ln)] = rs, 0.0
for (fl,ln),rs in fl_rolling_sums[1:]:
    corr = max(rs.corr(x) for x,_ in uncorrelated.values())
    if corr < CORR_THRESH:
        uncorrelated[(fl,ln)] = rs, corr
[(k, uncorrelated[k][1]) for k in uncorrelated.keys()]  
```

```python
uncorr_details = { (fl,ln) : (corr, df[(df.file==fl) & (df.line==ln)].file.count())
    for (fl,ln),(rs,corr) in uncorrelated.items()
}
uncorr_details
```

```python
uncorr_details_list = sorted(uncorr_details.items(), key=lambda x: -x[1][1])
uncorr_details_list
```

```python
np.set_printoptions(linewidth=160)
correlations = np.zeros((len(fl_rolling_sums),len(fl_rolling_sums)))
for i,(_,rs1) in enumerate(fl_rolling_sums):
    for j,(_,rs2) in enumerate(fl_rolling_sums):
        correlations[i,j] = rs1.corr(rs2) 
correlations        
```

```python
np.abs(correlations) <= 0.8
```

```python
N = len(correlations)
i_j_corr = [(i,j,correlations[i,j]) for i in range(N) for j in range(N) if i < j]
i_j_corr.sort(key=lambda x: (x[2],x[0],x[1]))
i_j_corr[:10]
```

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(correlations[:70])
print pca.explained_variance_ratio_
print pca.explained_variance_ratio_.cumsum()
```

### The file:line breakdown of the top 2 principal components.

```python
components = {
  'file:line': ['%s:%d' % (fl,ln) for fl,ln in fl_sorted.index.values[:NUM_FL][:70]],
}
for i in range(3):
    components['pc%d' % i] = pca.components_.T[i]
for v in components.values()[1:]:
    assert len(v) == len(components.values()[0]), '\n%s\n%s' % (v,components.values()[0]) 


comp_df = DataFrame(components)   
comp_df['t01'] = comp_df.pc1 - comp_df.pc0
pd.set_printoptions(max_rows=200, max_columns=20, precision=4)
open('pca.html', 'wt').write(comp_df.to_html())
comp_df[:30]
   
```
